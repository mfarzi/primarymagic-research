# Data Exploration - Clean Data Analysis

This notebook analyzes the percentage of clean data across aminoacids, dipeptides, and tripeptides datasets.

```python
import numpy as np
from pathlib import Path
from IPython.display import display, HTML
```

```python
def analyze_folder(base_path):
    """Analyze clean_data.npz vs data.npz for all subfolders."""
    results = []
    base = Path(base_path)
    
    for subfolder in sorted(base.iterdir()):
        if not subfolder.is_dir():
            continue
            
        data_file = subfolder / 'data.npz'
        clean_file = subfolder / 'clean_data.npz'
        
        if not data_file.exists() or not clean_file.exists():
            continue
        
        # Load and count spectra
        data = np.load(data_file, allow_pickle=True)
        clean_data = np.load(clean_file, allow_pickle=True)
        
        total_spectra = data['intensities'].shape[0]
        clean_spectra = clean_data['intensities'].shape[0]
        percentage = (clean_spectra / total_spectra) * 100
        
        results.append({
            'name': subfolder.name,
            'total': total_spectra,
            'clean': clean_spectra,
            'percentage': percentage
        })
    
    # Sort by percentage (low to high)
    results.sort(key=lambda x: x['percentage'])
    return results

def display_table(results, title):
    """Display results as an HTML table."""
    html = f"<h3>{title}</h3>"
    html += "<table border='1' style='border-collapse: collapse; width: 100%;'>"
    html += "<tr style='background-color: #f0f0f0;'>"
    html += "<th style='padding: 8px; text-align: left;'>Name</th>"
    html += "<th style='padding: 8px; text-align: right;'>Total Spectra</th>"
    html += "<th style='padding: 8px; text-align: right;'>Clean Spectra</th>"
    html += "<th style='padding: 8px; text-align: right;'>Clean %</th>"
    html += "</tr>"
    
    for r in results:
        color = '#ffcccc' if r['clean'] < 100 else '#ffffcc' if r['clean'] < 200 else '#ccffcc'
        html += f"<tr style='background-color: {color};'>"
        html += f"<td style='padding: 8px;'>{r['name']}</td>"
        html += f"<td style='padding: 8px; text-align: right;'>{r['total']}</td>"
        html += f"<td style='padding: 8px; text-align: right;'>{r['clean']}</td>"
        html += f"<td style='padding: 8px; text-align: right;'>{r['percentage']:.2f}%</td>"
        html += "</tr>"
    
    html += "</table>"
    
    avg = sum(r['percentage'] for r in results) / len(results) if results else 0
    total_all = sum(r['total'] for r in results)
    clean_all = sum(r['clean'] for r in results)
    overall = (clean_all / total_all * 100) if total_all > 0 else 0
    
    html += f"<p><b>Subfolders:</b> {len(results)} | "
    html += f"<b>Average Clean:</b> {avg:.2f}% | "
    html += f"<b>Overall Clean:</b> {clean_all}/{total_all} ({overall:.2f}%)</p>"
    
    display(HTML(html))
    return {'count': len(results), 'total': total_all, 'clean': clean_all, 'avg': avg, 'overall': overall}
```

## 1. Aminoacids

```python
aminoacids_results = analyze_folder('./data/processed/aminoacids')
aminoacids_stats = display_table(aminoacids_results, 'Aminoacids - Clean Data Analysis')
```

## 2. Dipeptides

```python
dipeptides_results = analyze_folder('./data/processed/dipeptides')
dipeptides_stats = display_table(dipeptides_results, 'Dipeptides - Clean Data Analysis')
```

## 3. Tripeptides

```python
tripeptides_results = analyze_folder('../data/processed/tripeptides')
tripeptides_stats = display_table(tripeptides_results, 'Tripeptides - Clean Data Analysis')
```

## Summary

```python
summary_html = "<h3>Overall Summary</h3>"
summary_html += "<table border='1' style='border-collapse: collapse; width: 100%;'>"
summary_html += "<tr style='background-color: #f0f0f0;'>"
summary_html += "<th style='padding: 8px;'>Category</th>"
summary_html += "<th style='padding: 8px;'>Subfolders</th>"
summary_html += "<th style='padding: 8px;'>Total Spectra</th>"
summary_html += "<th style='padding: 8px;'>Clean Spectra</th>"
summary_html += "<th style='padding: 8px;'>Avg Clean %</th>"
summary_html += "<th style='padding: 8px;'>Overall Clean %</th>"
summary_html += "</tr>"

for name, stats in [('Aminoacids', aminoacids_stats), ('Dipeptides', dipeptides_stats), ('Tripeptides', tripeptides_stats)]:
    summary_html += f"<tr>"
    summary_html += f"<td style='padding: 8px;'>{name}</td>"
    summary_html += f"<td style='padding: 8px; text-align: center;'>{stats['count']}</td>"
    summary_html += f"<td style='padding: 8px; text-align: right;'>{stats['total']}</td>"
    summary_html += f"<td style='padding: 8px; text-align: right;'>{stats['clean']}</td>"
    summary_html += f"<td style='padding: 8px; text-align: right;'>{stats['avg']:.2f}%</td>"
    summary_html += f"<td style='padding: 8px; text-align: right;'>{stats['overall']:.2f}%</td>"
    summary_html += "</tr>"

summary_html += "</table>"
display(HTML(summary_html))
```
